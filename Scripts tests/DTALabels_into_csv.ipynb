This notebook gets all labels from dta files for each database and export the dict into csv files.

```python
import csv 
#Importer les packages 
import pandas as pd
from pathlib import Path
import os
import csv
import glob
```

```python
#indicate folder 
dta_directory = Path("C:/Users/user/Documents/Municipales_2026/Données/dta_files")
output_directory =Path("C:/Users/user/Documents/Municipales_2026/Output")
dta_directory
```

```python
#ouvrir un fichier dta pour obtenir les labels

itr = pd.read_stata('', iterator=True)
glossaire_alphabetisation=itr.variable_labels()
type(glossaire_alphabetisation)
```

```python


for file in dta_directory.glob("*.dta"):
    itr = pd.read_stata(file, iterator=True)
    file_name=file.stem
    labels_dict=itr.variable_labels()
    
    with open("dict_"+file_name+".csv", "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
    
        # en-tête
        writer.writerow(["terme", "definition"])
    
        # une ligne par entrée du dictionnaire
        for terme, definition in labels_dict.items():
            writer.writerow([terme, definition])





```
